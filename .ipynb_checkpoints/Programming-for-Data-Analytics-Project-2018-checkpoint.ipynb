#### Chosen problem: OBESITY
#### Population: female, adults
#### Country: Ireland

[Obesity] <img src="https://i.ibb.co/M2ZW7bw/scale-2396062-1920.jpg" alt="scale-2396062-1920" border="0"  width="25%">

In Ireland 60% of adults are overweight (excess weight or obese) [1]. Besides, the prevalence of excess weight/obesity in adults from 18 to 64 years in Ireland has increased since 1990. Therefore, it still is a major public health problem in Ireland [2].
It's estimated that an adult woman needs 1600 to 2400 calories per day and it depends on a number of factors, such as age, height, weight and level of physical activity. So, the low end of the calories needed per day is for sedentary women and the higher end of the range is for the active women [3].
It is well known that to maintain a healthy weight the amount of calories consumed must be in balance with the amount of calories burnt. Therefore, the amount of calories consumed must be greater than the calories burnt to result in  a gain weight [4].
It is no wonder that  the average daily consumption of calories in Ireland is 2307 calories. The world  average daily consumption of calories is 1398. So, come as no surprise that the average daily energy consumption of calories in Ireland it is one of the biggest in the world [5]. 

The dataset created includes only overweight and obese adult women in Ireland. 

#### The variables chosen are: 

1.  Age
2. BMI (Body Mass Index)
3. Energy intake per day
4. WHR (Waist:Hip Ratio)


#### Age: 
The average age of adults in Ireland is 37.4 years [6]. Surveys in Ireland usually include adults from 18 to 64 years old.  

#### BMI:
The BMI (Quetelet Index) indicates the nutritional status in adults and it's a measure defined as the weight (in kilograms) divided by the square of height (in metres) [9].The World Health Organisation (WHO) BMI cut-off points were used to estimate the nutritional status. As there is no maximum BMI for obese women, it was considered 45 kg/m2 as the BMI of morbidly obese person is equal or greater than 40 kg/m2. 



[BMI formula] <img src="https://image.ibb.co/io1FqL/BMI-Formula-1.jpg" alt="BMI-Formula-1" border="0" width="30%">



[BMI nutritional status] <img src="https://image.ibb.co/fggtCA/BMI-chart.png" alt="BMI-chart" border="0" width="50%"> 


The National Adult Nutrition Survey in 2011 with a 1500 adults (740 men and 760 women) showed that among 18 to 64 years old women a total of 21% were obese and 31% were overweight [2]. Two lists of random BMI values were created based on NANS 2011 proportion of number of women overweight and obese.


#### Energy intake per day: 
The average consumption of calories in Ireland is 2307 [5].

#### WHR (Waist:Hip Ratio):
The adult Irish population has an average value of 0.84 (std 0.08) of WHR [2]. As body fat distribution is an important indicator of disease risk, the WHR is used to predict the risk of a cardiovascular disease (CVD). It was showed an increased WHR in women, which indicated a high risk of CVD [2]. The WHO cut off for WHR for increased risk of CVD is 0.85 (as showed in the following table).


[WHR WHO cut off] <img src="https://i.ibb.co/tBvwyD1/whrwhocutoff.png" alt="whrwhocutoff" border="0" width="90%">



#### Number of the sample of the project: 520 women

Based on NANS 2011 Survey with a sample of 1500 individuals (760 women), where 48% of them had healthy weight, 31% were overweight and 21% were obese.
I've decided to get just the data from women with any degree of excess weight. Therefore, among 1000 women (48% healthy weight, 31% overweight, 21% obese), 480 would have a healthy weight and 520 would have a degree of excess of weight (310 overweight and 210 obese).

#### Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns; sns.set()
```

### Age

##### Generate random values of age using numpy.random function

```python
# Generate age between 18 and 64 years old

age = np.random.randint(18, 64, 520)

age
```

##### Age - check the numbers generated : mean, median, plots (histogram and boxplot)

```python
# Average age of the sample

np.mean(age)
```

* Mean higher than the national average of 37.4 years old

```python
# Median 

np.median(age)
```

```python
#Plot the age

# Histogram

plt.figure(figsize=(12,8), facecolor='w', edgecolor='k', clear='True')
plt.hist(age, bins=100)
plt.xlabel('Age in years')
plt.show()
```

* Even generating numbers between 18 to 64, it is clear that the main peaks are around 21 and 46 years.

```python
#Boxplot

plt.boxplot(age, showmeans=True)
plt.show()
```

* Red triangle is the mean and the line inside the box is the median. Normal distribution - mean similar to median.

### Body Mass Index (BMI)

#### Generate random BMI values for 2 groups of women: overweight and obese

##### Overweight - generate BMI values using numpy.random function

```python
# Overweight - BMI greater or equal 25kg/m2 and less than 30 kg/m2

#Min and max weight to be classified as overweight

min_bmi_overweight = 25
max_bmi_overweight = 29.99
```

```python
bmi_overweight = np.random.uniform(25,29.99, 310)

bmi_overweight
```

##### Overweight /BMI - check the numbers generated: average, median, plots (histogram and boxplot)

```python
# Average BMI of overweight women

np.mean (bmi_overweight)
```

```python
np.median(bmi_overweight)
```

```python
#Plot overweight sample

# Histogram

plt.figure(figsize=(12,8), facecolor='w', edgecolor='k', clear='True')
plt.hist(bmi_overweight)
plt.show()
```

* Peak of BMI of overweight women is at values between 27 and 28.

```python
#Box plot

plt.boxplot(bmi_overweight, showmeans=True)
plt.show()
```

* Normal distribution - mean similar to median. Large spread of the data as the IQR is big (from 26 kg/m2 to 28.5 kg/m2).

##### Obese - generate BMI values using numpy.random function

```python
# obesity - BMI greater or equal 30 kg/m2 (there is no max BMI, but it was considered 45 kg/m2)

#Min and 'max' BMI to be classified as obese

min_bmi_obese = 30
max_bmi_obese = 45
```

```python
bmi_obese = np.random.uniform(30,45, 210)

bmi_obese
```

##### Obese /BMI - check the numbers generated: average, median, plots (histogram and boxplot)

```python
# Average and median BMI of obese women

np.mean (bmi_obese)
```

```python
np.median (bmi_obese)
```

```python
#Plot overweight sample

# Histogram

plt.figure(figsize=(12,8), facecolor='w', edgecolor='k', clear='True')
plt.hist(bmi_obese)
plt.show()
```

* Peaks of BMI of obese women at values between 30 and 31 and between 41 and 42.

```python
# Box plot

plt.boxplot(bmi_obese, showmeans=True)
plt.show()
```

* The data is slightly skewed to the left. Data is spread, as the "box" (Interquartile range - IQR) is big (from 33 kg/m2 to 41 kg/m2).

##### Create one single array containing all the sample BMI (overweight and obese)

```python
bmi = np.concatenate ((bmi_overweight, bmi_obese), axis=None)
bmi
```

##### BMI (overweight & obese) - check the numbers generated: average, median, plots (histogram and boxplot)

```python
# Average BMI of the sample

np.mean(bmi)
```

```python
# Median BMI 

np.median(bmi)
```

```python
# Plot - BMI

# Histogram

plt.figure(figsize=(12,8), facecolor='w', edgecolor='k', clear='True')
plt.hist(bmi, bins=50)
plt.xlabel('BMI kg/m2')
plt.show()
```

* BMI is right-skewed.
* Main peaks of BMI: overweight around 25.5 kg/m2 and obese around 41 kg/m2.

```python
# Boxplot

plt.boxplot(bmi, showmeans=True)
plt.show()
```

* Mean greater than the median - skewed to the right with a long tail of high BMI.

### Daily energy intake

###### Generate random daily energy intake values using numpy.random function

```python
energy_intake = np.random.normal (2307, 500, 520)

energy_intake
```

##### Daily energy intake - check the numbers generated: min/max, average, median, plots (histogram and boxplot)

```python
#Maximum and minimun daily energy intake

np.amax(energy_intake)
```

```python
np.amin(energy_intake)
```

* Energy intake from is 339.64 to 3679.73 kcal per day. Probably there is underreporting of energy consumed by individual from the sample or some of them might be on a diet. Further investigation is needed.

```python
#Median and average energy intake per day

np.median(energy_intake)
```

```python
np.mean(energy_intake)
```

```python
# Plot - Daily Energy Intake

# Histogram

plt.figure(figsize=(12,8), facecolor='w', edgecolor='k', clear='True')
plt.hist(energy_intake, bins=50)
plt.xlabel('Daily energy intake (kcal)')
plt.show()
```

* Histogram has nearly a Bell shape indicating a normal distribution.
* Data is widely spread: from less than 500 kcal to greater than 3500 kcal.
* Main peak of energy intake per day is at 2400 kcal.

```python
#Box plot

plt.boxplot (energy_intake, showfliers=True, showmeans=True)
plt.show()
```

* Normal distribution - mean similar to median. Daily Energy Intake between 1000 kcal and 3500 kcal. The data is not so spread as its IQR is small (from 2000 kcal to 2500 kcal).

### Waist - Hip Ratio (WHR)

##### Generate Waist:Hip Ratio - WHR numbers using numpy.random function

```python
whr = np.random.normal(0.84, 0.08, 520)

whr
```

##### WHR - check the numbers generated: min/max, average, median, plots (histogram and boxplot)

```python
# Max and min WHR value

np.amax(whr)
```

```python
np.amin(whr)
```

* WHR values from 0.61 to 1.09. Individual with WHR of 0.61 might present a body shape of 'pear' (hip bigger than waist).

```python
# Plot - WHR

#Histogram

#Main peak at 0.85

plt.figure(figsize=(12,8), facecolor='w', edgecolor='k', clear='True')
plt.hist(whr, bins=50)
plt.xlabel('Waist-Hip Ratio')
plt.show()
```

```python
#Boxplot

#Normal distribution - mean similar to median

plt.boxplot (whr, showmeans=True)
plt.show ()
```

### Dataframe containing all data generated

```python
#Create dataframe with all the data randomly generated

df=pd.DataFrame(data={'Age': age, 'BMI': bmi, 'WHR': whr, 'Daily Energy Intake': energy_intake})

# Add a column named NutritionalStatus - if the BMI is less than 30 kg/m2 the woman is classified as overweight, otherwise is classified as obese.

df['NutritionalStatus'] = np.where(df['BMI'] < 30,'Overweight','Obese')

# Check the first 20 rows of the dataframe
df.head(20)
```

##### Summarized view of the data frame

```python
df.describe()
```

#####  Check distribution between Nutritional Status and other variables

```python
# Nutritional Status vs Age

sns.set_style("whitegrid")
sns.boxplot(x="NutritionalStatus", y="Age", data = df)
sns.despine()
plt.show()
```

* Mean, median, hinge spread, the right and the left whiskers of both groups (overweight vs obese women) are quite similar regarding the age of them.

```python
# Nutritional Status vs Daily Energy Intake

sns.set_style("whitegrid")
sns.boxplot(x="NutritionalStatus", y="Daily Energy Intake", data = df)
sns.despine()
plt.show()
```

* Overweight women have more outliers values less than 1000 kcal consumed per day. Basically the mean, median and both whiskers are similar.

```python
# Nutritional Status vs WHR

sns.set_style("whitegrid")
sns.boxplot(x="NutritionalStatus", y="WHR", data = df)
sns.despine()
plt.show()
```

* Obese women showed an outlier value greater than 1 and overweight women have outliers less than 0.7. The mean and the median of both groups of women are similar, but obese women showed the left and right whiskers longer than the overweight women.

##### Observe the relationship between variables based on Nutritional Status

```python
# Create pairplot of the dataframe

sns.pairplot(df, hue="NutritionalStatus", kind="reg")
plt.show()
```

#### The graphs above show:
* It is clear that there are 2 distinct groups by looking at the scatter plot of BMI: overweight and obese women.
* Obese women: the smaller the age the higher is the BMI.
* Overweight women: the higher the BMI the bigger is the daily energy intake and the WHR.
* Bell shape/normal distribution of following histograms: Daily energy intake and WHR.
* Uniform distribution: age.
