#### Chosen problem: OBESITY
#### Population: female, adults
#### Country: Ireland

<img src="https://i.ibb.co/M2ZW7bw/scale-2396062-1920.jpg" alt="scale-2396062-1920" border="0"  width="25%">

In Ireland 60% of adults are overweight (excess weight or obese) [1]. Besides, the prevalence of excess weight/obesity in adults from 18 to 64 years in Ireland has increased since 1990. Therefore, it still is a major public health problem in Ireland [2].
It's estimated that an adult woman needs 1600 to 2400 calories per day and it depends on a number of factors, such as age, height, weight and level of physical activity. So, the low end of the calories needed per day is for sedentary women and the higher end of the range is for the active women [3].
It is well known that to maintain a healthy weight the amount of calories consumed must be in balance with the amount of calories burnt. Therefore, the amount of calories consumed must be greater than the calories burnt to result in  a gain weight [4].
It is no wonder that  the average daily consumption of calories in Ireland is 2307 calories. The world  average daily consumption of calories is 1398. So, come as no surprise that the average daily energy consumption of calories in Ireland it is one of the biggest in the world [5]. 

The database created includes only overweight and obese adult women in Ireland. 

#### The variables chosen are: 

1.  Age
2. BMI (Body Mass Index)
3. Energy intake per day
4. WHR (Waist:Hip Ratio)


#### Age: 
The average age of adults in Ireland is 37.4 years [6]. Surveys in Ireland usually include adults from 18 to 64 years old.  

#### BMI:
The BMI (Quetelet Index) indicates the nutritional status in adults and it's a measure defined as the weight (in kilograms) divided by the square of height (in metres) [9].The World Health Organisation (WHO) BMI cut-off points were used to estimate the nutritional status. As there is no maximum BMI for obese women, it was considered 45 kg/m2 as the BMI of morbidly obese person is equal or greater than 40 kg/m2.

The National Adult Nutrition Survey in 2011 with a 1500 adults (740 men and 760 women) showed that among 18 to 64 years old women a total of 21% were obese and 31% were overweight [2]. Two lists of random BMI values were created based on NANS 2011 proportion of number of women overweight and obese.


#### Energy intake per day: 
The average consumption of calories in Ireland is 2307 [5].

#### WHR (Waist:Hip Ratio):
The adult Irish population has an average value of 0.84 (std 0.08) of WHR [2]. As body fat distribution is an important indicator of disease risk, the WHR is used to predict the risk of a cardiovascular disease (CVD). It was showed an increased WHR in women, which indicated a high risk of CVD [2]. The WHO cut off for WHR for increased risk of CVD is 0.85 (as showed in the following table).


<img src="https://i.ibb.co/tBvwyD1/whrwhocutoff.png" alt="whrwhocutoff" border="0" width="90%">



#### Number of the sample of the project: 520 women

Based on NANS 2011 Survey with a sample of 1500 individuals (760 women), where 48% of them had healthy weight, 31% were overweight and 21% were obese.
I've decided to get just the data from women with any degree of excess weight. Therefore, among 1000 women (48% healthy weight, 31% overweight, 21% obese), 480 would have a healthy weight and 520 would have a degree of excess of weight (310 overweight and 210 obese).

<center><img src="https://i.ibb.co/HtQGXHD/samplesizeproject.png" alt="samplesizeproject" border="0"></center>


```python
#Import libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns; sns.set()
```

### Age

```python
# Generate age between 18 and 64 years old

age = np.random.randint(18, 64, 520)

age
```

```python
# Average age of the sample

# Slightly higher than the national average of 37.4 years old

np.mean(age)
```

```python
# Median 

np.median(age)
```

```python
#Plot the age

# Histogram

# 3 main peaks: 32, 46, 26

plt.figure(figsize=(12,8), facecolor='w', edgecolor='k', clear='True')
plt.hist(age, bins=100)
plt.xlabel('Age in years')
plt.show()
```

```python
#Boxplot

plt.boxplot(age, showmeans=True)
plt.show()
```

### Body Mass Index (BMI)

```python
# Overweight - BMI greater or equal 25kg/m2 and less than 30 kg/m2

#Min and max weight to be classified as overweight

min_bmi_overweight = 25
max_bmi_overweight = 29.99
```

```python
bmi_overweight = np.random.uniform(25,29.99, 310)

bmi_overweight
```

```python
# Average BMI of overweight women

np.mean (bmi_overweight)
```

```python
np.median(bmi_overweight)
```

```python
#Plot overweight sample

# Histogram

# Peaks at values between 28 and 29 and at values between 29 and 29.99

plt.figure(figsize=(12,8), facecolor='w', edgecolor='k', clear='True')
plt.hist(bmi_overweight)
plt.show()
```

```python
#Boxplt

plt.boxplot(bmi_overweight, showmeans=True)
plt.show()
```

```python
# obesity - BMI greater or equal 30 kg/m2 (there is no max BMI, but it was considered 45 kg/m2)

#Min and 'max' BMI to be classified as obese

min_bmi_obese = 30
max_bmi_obese = 45
```

```python
bmi_obese = np.random.uniform(30,45, 210)

bmi_obese
```

```python
# Average BMI of obese women

np.mean (bmi_obese)
```

```python
np.median (bmi_obese)
```

```python
#Plot overweight sample

# Histogram

# Peaks at values between 30 and 31 and at 43

plt.figure(figsize=(12,8), facecolor='w', edgecolor='k', clear='True')
plt.hist(bmi_obese)
plt.show()
```

```python
plt.boxplot(bmi_obese, showmeans=True)
plt.show()
```

```python
#Create one single array containing all the sample BMI (overweight and obese)

bmi = np.concatenate ((bmi_overweight, bmi_obese), axis=None)
bmi
```

```python
# Average BMI of the sample

np.mean(bmi)
```

```python
np.median(bmi)
```

```python
# Plot - BMI

# Histogram

# Main peaks: at 25 and between 28 and 29 for overweight and at 37 and at 42 for obese.

plt.figure(figsize=(12,8), facecolor='w', edgecolor='k', clear='True')
plt.hist(bmi, bins=100)
plt.xlabel('BMI kg/m2')
plt.show()
```

```python
# Boxplot

plt.boxplot(bmi, showmeans=True)
plt.show()
```

### Daily energy intake

```python
# Generate daily energy intake values

energy_intake = np.random.normal (2307, 500, 520)

energy_intake
```

```python
np.amax(energy_intake)
```

```python
np.amin(energy_intake)
```

```python
# Energy intake from 784 to 3581 kcal per day.
```

```python
# Plot - Daily Energy Intake

# Histogram

Main peaks: at 2300 and at value between 2500 and 3000

plt.figure(figsize=(12,8), facecolor='w', edgecolor='k', clear='True')
plt.hist(energy_intake, bins=100)
plt.xlabel('Daily energy intake (kcal)')
plt.show()
```

```python
plt.boxplot (energy_intake, showfliers=True, showmeans=True)
plt.show()
```

### Waist - Hip Ratio (WHR)

```python
# Generate Waist:Hip Ratio - WHR numbers

whr = np.random.normal(0.84, 0.08, 520)

whr
```

```python
np.amax(whr)
```

```python
np.amin(whr)

#WHR values from 0.60 to 1.09
```

```python
# Plot - WHR

#Histogram

plt.figure(figsize=(12,8), facecolor='w', edgecolor='k', clear='True')
plt.hist(whr, bins=100)
plt.xlabel('Waist-Hip Ratio')
plt.show()
```

```python
#Boxplot

plt.boxplot (whr, showmeans=True)
plt.show ()
```

### Dataframe containing all data generated

```python
#Create dataframe with all the data randomly generated

df=pd.DataFrame(data={'Age': age, 'BMI': bmi, 'WHR': whr, 'Daily Energy Intake': energy_intake})

df
```

```python
# Create pairplot of the dataframe

sns.pairplot(df)
plt.show()
```

```python
d = sns.PairGrid(df)
d.map(plt.scatter)
plt.show()
```
