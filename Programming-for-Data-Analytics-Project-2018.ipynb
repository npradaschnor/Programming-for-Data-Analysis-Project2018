In Ireland 60% of adults are overweight (excess weight or obese) [1]. Besides, the prevalence of excess weight/obesity in adults from 18 to 64 years in Ireland has increased since 1990. Therefore, it still is a major public health problem in Ireland [2].
It's estimated that an adult woman needs 1600 to 2400 calories per day and it depends on a number of factors, such as age, height, weight and level of physical activity. So, the low end of the calories needed per day is for sedentary women and the higher end of the range is for the active women [3].
It is well known that to maintain a healthy weight the amount of calories consumed must be in balance with the amount of calories burnt. Therefore, the amount of calories consumed must be greater than the calories burnt to result in  a gain weight [4].
It is no wonder that  the average daily consumption of calories in Ireland is 2307 calories. The world  average daily consumption of calories is 1398. So, come as no surprise that the average daily energy consumption of calories in Ireland it is one of the biggest in the world [5]. 

The database that created includes only overweight and obese adult women in Ireland. The variables chosen are: 

1.  Age
2. BMI (Body Mass Index)
3. Energy intake per day
4. WHR (Waist:Hip Ratio)


#### Age: 
The average age of adults in Ireland is 37.4 years [6]. Surveys in Ireland usually include adults from 18 to 64 years old.  

#### BMI:
The BMI (Quetelet index) indicates the nutritional status in adults and it's a measure defined as the weight (in kilograms) divided by the square of height (in metres) [9].The World Health Organisation (WHO) BMI cut-off points were used to estimate the weight. As there is no maximum weight to be obese it was considered a a 'max' BMI of superobesity (50kg/m2).

The National Adult Nutrition Survey in 2011 with a 1500 adults (740 men and 760 women) showed that among 18 to 64 years old women a total of 21% were obese and 31% were overweight [2]. Two lists of random BMI values were created based on NANS 2011 proportion of number of women overweight and obese.


#### Energy intake per day: 
The average consumption of calories in Ireland is 2307 [5].

#### WHR (Waist:Hip Ratio):
The adult irish population has an average value of 0.84 (std 0.08) of WHR [2]. As body fat distribution is an important indicator of disease risk, the WHR is used to predict the risk of a cardiovascular disease (CVD). It was showed an increased WHR in women, which indicated a high risk of CVD [2]. The WHO cut off for WHR for increased risk of CVD is 0.85 (as showed in the following table).

#### Number of the sample of the project: 
Based on NANS 2011 Survey with a sample of 1500 individuals (760 women), where 48% healthy weight, 31% overweight, 21% obese, I've decided to get just the data from women with any degree of excess weight. Therefore, among 1000 women (48% healthy weight, 31% overweight, 21% obese), 480 would have a healthy weight and 520 would have a degree of excess of weight (310 overweight and 210 obese).


```python
#Import libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns; sns.set()

# Generate age between 18 and 64 years old

age = np.random.randint(18, 64, 520)

age
```

```python
#Plot the age

plt.hist(age, bins=50)
plt.xlabel('Age in years')
plt.show()
```

```python
# overweight - BMI greater or equal 25kg/m2 and less than 30 kg/m2

#Min and max weight to be classified as overweight

min_bmi_overweight = 25
max_bmi_overweight = 29.99
```

```python
bmi_overweight = np.random.uniform(25,29.99, 310)

bmi_overweight
```

```python
# obesity - BMI greater or equal 30kg/m2 (there is no max BMI, but it was considered 50kg/m2)

#Min and 'max' weight to be classified as obese

min_bmi_obese = 30
max_bmi_obese = 50
```

```python
bmi_obese = np.random.uniform(30,50, 210)

bmi_obese
```

```python
bmi = np.concatenate ((bmi_overweight, bmi_obese), axis=None)
bmi
```

```python
# Plot - BMI

plt.hist(bmi, bins=100)
plt.xlabel('BMI kg/m2')
plt.show()
```

```python
plt.boxplot(bmi)
plt.show()
```

```python
# Generate daily energy intake

energy_intake = np.random.normal (2307, 500, 520)

energy_intake
```

```python
# Daily Energy Intake - Plot

plt.hist(energy_intake, bins=100)
plt.xlabel('Daily energy intake (kcal)')
plt.show()
```

```python
plt.boxplot (energy_intake)
plt.show()
```

```python
# Generate Waist:Hip Ratio - WHR

whr = np.random.normal(0.84, 0.08, 520)

whr
```

```python
plt.boxplot (whr)
plt.show ()
```

```python
#Create dataframe with all the data randomly generated

df=pd.DataFrame(data={'Age': age, 'BMI': bmi, 'WHR': whr, 'Daily Energy Intake': energy_intake})

df
```

```python
# Create pairplot

sns.pairplot(df)
plt.show()
```

```python
d = sns.PairGrid(df)
d.map(plt.scatter)
plt.show()
```

```python
# Plot - Regression age x daily energy intake

sns.lmplot(x="Age", y="Daily Energy Intake", data=df)
plt.show()
```

```python
# Plot - Regression age x BMI

sns.lmplot(x="Age", y="BMI", data=df)
plt.show()
```

```python
# Plot - Regression BMI x WHR

sns.lmplot(x="BMI", y="WHR", data=df)
plt.show()
```
