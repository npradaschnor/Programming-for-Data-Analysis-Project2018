#### Country: Ireland

In Ireland 60% of adults are overweight (excess weight or obese) [1]. Besides, the prevalence of excess weight/obesity in adults from 18 to 64 years in Ireland has increased since 1990. Therefore, it still is a major public health problem in Ireland [2].
It's estimated that an adult woman needs 1600 to 2400 calories per day and it depends on a number of factors, such as age, height, weight and level of physical activity. So, the low end of the calories needed per day is for sedentary women and the higher end of the range is for the active women [3].
It is well known that to maintain a healthy weight the amount of calories consumed must be in balance with the amount of calories burnt. Therefore, the amount of calories consumed must be greater than the calories burnt to result in  a gain weight [4].
It is no wonder that  the average daily consumption of calories in Ireland is 2307 calories. The world  average daily consumption of calories is 1398. So, come as no surprise that the average daily energy consumption of calories in Ireland it is one of the biggest in the world [5]. 

The database that created includes only overweight and obese adult women in Ireland. The variables chosen are:

#### Number of the sample of the project: 
Based on the NANS 2011 Survey with a sample of 1500 individuals (760 women), where 48% healthy weight, 31% overweight, 21% obese, I've decided to get just the data from women with any degree of excess weight. Therefore, among 1000 women (48% healthy weight, 31% overweight, 21% obese), 480 would have a healthy weight and 520 would have a degree of excess of weight.


```python
#Import libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Generate age between 18 and 64 years old

age = np.random.randint(18, 64, 520)

age
```

```python
#Plot the age

plt.hist(age, bins=50)
plt.xlabel('Age in years')
plt.show()
```

```python
# Calculate weight (min and max) of the sample

# Square of height (in m)
h = 1.63

h2 = h * h

h2
```

```python
# overweight - BMI greater or equal 25kg/m2 
min_bmi = 25

min_weight = h2 * min_bmi

min_weight
```

```python
# superobesity - BMI greater or equal 50kg/

max_bmi = 50

max_weight = h2 * max_bmi

max_weight
```

```python
# Generate random height

height= np.random.normal(1.63, 0.01, 520)

height
```

```python
# Height - Plot
```

```python
# Generate random weight

weight = np.random.randint(66.4225, 132.9, 520)

weight
```

```python
# Weight - Plot
```

```python
# Calculate BMI of the sample

h2 = height * height

h2
```

```python
bmi = weight/h2

bmi
```

```python
# Plot - BMI

plt.hist(bmi, bins=50)
plt.xlabel('BMI kg/m2')
plt.show()
```

```python
overweight = bmi < 30 

overweight
```

```python
total_number_sample = 520
```

```python
# Get the number of overweight women

overweight_number = np.sum (overweight)

overweight_number
```

```python
overweight_per_cent = (overweight_number * 100)/total_number_sample

overweight_per_cent
```

```python
# Get the number of obese women

obese_number = total_number_sample - overweight_number

obese_number
```

```python
obese_per_cent = (obese_number * 100)/total_number_sample

obese_per_cent
```

```python
#Obese subgroup - degree of obesity (from I to III)
```

```python
#Number of obese I women

obese_one_number = np.sum((bmi >= 30 ) & (bmi < 35))

obese_one_number
```

```python
#Per cent of obese I women (from the total number of obese women)

obese_one_number_per_cent = obese_one_number * 100/obese_number

obese_one_number_per_cent 
```

```python
#Number of obese II women

obese_two_number = np.sum((bmi >= 35 ) & (bmi < 40))

obese_two_number
```

```python
#Per cent of obese II women (from the total number of obese women)

obese_two_number_per_cent = obese_two_number * 100/obese_number

obese_two_number_per_cent 
```

```python
#Number of obese III women

obese_three_number = np.sum(bmi >= 40)

obese_three_number
```

```python
#Per cent of obese III women (from the total number of obese women)

obese_three_number_per_cent = obese_three_number * 100/obese_number

obese_three_number_per_cent 
```

```python
#Age x BMI

plt.scatter(age, bmi)
plt.show()
```

```python
# Generate daily energy intake

energy_intake = np.random.normal (2307, 500, 520)

energy_intake
```

```python
# Daily Energy Intake - Plot
```

```python
#Create dataframe with all the data randomly generated

df=pd.DataFrame(data={'Age': age, 'Height (m)': height, 'Weight (kg)': weight, 'BMI': bmi, 'Daily Energy Intake': energy_intake})

df
```

```python
# Create pairplot

sns.pairplot(df)
plt.show()
```
