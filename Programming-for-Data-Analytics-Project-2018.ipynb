In Ireland 60% of adults are overweight (excess weight or obese) [1]. Besides, the prevalence of excess weight/obesity in adults from 18 to 64 years in Ireland has increased since 1990. Therefore, it still is a major public health problem in Ireland [2].
It's estimated that an adult woman needs 1600 to 2400 calories per day and it depends on a number of factors, such as age, height, weight and level of physical activity. So, the low end of the calories needed per day is for sedentary women and the higher end of the range is for the active women [3].
It is well known that to maintain a healthy weight the amount of calories consumed must be in balance with the amount of calories burnt. Therefore, the amount of calories consumed must be greater than the calories burnt to result in  a gain weight [4].
It is no wonder that  the average daily consumption of calories in Ireland is 2307 calories. The world  average daily consumption of calories is 1398. So, come as no surprise that the average daily energy consumption of calories in Ireland it is one of the biggest in the world [5].
